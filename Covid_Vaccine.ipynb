```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt

f="COVID.xlsx"
df = pd.read_excel(f)
df
```

```python
df.describe()
```

```python
df['Black %'] = df['Black']/df['Black total']
```

```python
df2 = df.loc[df['Age Range'] == '80+']
df2['Año'] = df2['Covid Vacc Date'].dt.year
df2['Mes'] = df2['Covid Vacc Date'].dt.month
df2
```

```python
df3 = df2.groupby(['Año','Mes']).max()
df3
```

```python
df3['Black %'] = (df3['Black']/df3['Black total'])*100
df3['Mixed %'] = (df3['Mixed']/df3['Mixed total'])*100
df3['Other %'] = (df3['Other']/df3['Other total'])*100
df3['South Asian %'] = (df3['South Asian']/df3['South Asian total'])*100
df3['Unknown %'] = (df3['Unknown']/df3['Unknown total'])*100
df3['White %'] = (df3['White']/df3['White total'])*100
df3 = df3.reset_index()
df3
#df4 = df3.loc[df3['Mes'] == '2021']
#df4
```

```python
df4 = df3.loc[df3['Año'] == 2021]
df4
#df3
```

```python
plt.style.use('seaborn')
plt.figure(dpi=100) 
plt.plot(df4['Mes'],df4['Black %'])
plt.plot(df4['Mes'],df4['Mixed %'])
plt.plot(df4['Mes'],df4['Other %'])
plt.plot(df4['Mes'],df4['South Asian %'])
plt.plot(df4['Mes'],df4['Unknown %'])
plt.plot(df4['Mes'],df4['White %'])

plt.legend(['Black %', 'Mixed %', 'Other %', 'South Asian %', 'Unknown %', 'White %'])
plt.xlabel('Mes')
plt.ylabel('% COVID')
plt.title('% de personas vacunadas segun su etnia')
plt.show()
```
